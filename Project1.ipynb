```python
import pandas as pd
missing_values = ["s", "na"]
df = pd.read_csv('https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/master/2014-15_To_2016-17_School-_Level_NYC_Regents_Report_For_All_Variables.csv',
                na_values = missing_values)
# Cleaning data
# creating and mutating the dataframe where the values "s" and "na" are replaced with "NaN"
```

```python
# Dropping two columns that is nonsensical
drop_cols = ['Number Scoring CR', 'Percent Scoring CR']
df.drop(drop_cols, inplace = True, axis = 1)
```

```python
# dropping indexes that have more than 7 "NaN"
df.dropna(axis=0, thresh=7, inplace=True)
```

```python
# Used a bar plot to see the first 20 indexes Total tested from schools
import matplotlib.pyplot as plt
plt.bar(df['School Name'].head(20), df['Total Tested'].head(20))
plt.xticks(rotation=90)
plt.show()
```

```python
# Using groupby to compare the mean score for each school
school_means = df.groupby('School Name')['Mean Score'].mean()
school_means
```

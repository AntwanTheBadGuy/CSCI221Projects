```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('https://raw.githubusercontent.com/niteen11/DataAnalyticsAcademy/master/Python/dataset_diabetes/diabetic_data.csv')
```

```python
#Taking a quick insight of the data frame
print("Dataset shape:", df.shape)
print("\nColumn names:\n", df.columns.tolist())
df.head()
```

```python
#Cleaning tasks
df = df[df['gender'] != 'Unknown/Invalid']
print("\nAfter cleaning, shape:", df.shape)
```

```python
#Setting up for comparisons
group_feature = 'gender'
compare_feature = 'num_medications'
```

```python
#Descriptive data
group_stats = df.groupby(group_feature)[compare_feature].describe()
print("\nDescriptive statistics by gender:\n", group_stats)
```

```python
#Visualization for comparing the number of medications by gender
plt.figure(figsize=(8,6))
#Using boxplot from seaborn
sns.boxplot(x=group_feature, y=compare_feature, data=df)
plt.title('Comparison of Number of Medications by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Medications')
plt.grid(True)
plt.show()
```
